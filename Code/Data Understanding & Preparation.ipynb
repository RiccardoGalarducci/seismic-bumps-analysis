## INTRODUCTION OF THE SEISMIC-BUMPS DATASET

The data describe the problem of high energy seismic bumps forecasting in longwalls coal mines. The goal is to assess if a coal mine is prone to rockburst causing seismic hazards or not.

https://archive.ics.uci.edu/ml/datasets/seismic-bumps

The data is organised in a form of an array, where the dataset attributes (aka features) are organised in columns, while the rows represent the summary statement about seismic activity in the rock mass within one shift (8 hours), aka instances.<br>
The dataset is composed by 2584 instances and 19 columns.

```python
c_shift = ['#17becf', '#1f77b4']
c_shift_r = ['#1f77b4', '#17becf']

fig, ax2 = plt.subplots(figsize=(10, 6), subplot_kw=dict(aspect="equal"))

wedges, texts = ax2.pie(sizes, wedgeprops=dict(width=0.5), startangle=-100, shadow = True, colors = c_shift)

bbox_props = dict(boxstyle="round4,pad=0.3", fc="w", ec="k", lw=0.0)
kw = dict(arrowprops=dict(arrowstyle="-", color='k'), bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax2.annotate(sizes[i], fontstyle = 'italic', fontsize = 16, xy=(x, y), xytext=(1.15*np.sign(x), 1.2*y), horizontalalignment=horizontalalignment, **kw)

ax2.legend(wedges, labels, title="Shift", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize = 14, title_fontsize = 12, facecolor='white')    
    
plt.show()
```

```python
sizes_total = sizes[0] + sizes[1]
sizes_perc = sizes.copy()
for i in range(0,2):
    sizes_perc[i] = sizes_perc[i] * 100 / sizes_total
sizes_perc
```

```python
fig, ax2 = plt.subplots(figsize=(10, 6), subplot_kw=dict(aspect="equal"))

wedges, texts = ax2.pie(sizes, wedgeprops=dict(width=0.5), startangle=-80, shadow = True, colors = c_shift)

bbox_props = dict(boxstyle="round4,pad=0.3", fc="w", ec="k", lw=0.0)
kw = dict(arrowprops=dict(arrowstyle="-", color='k'), bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    percentage = '{:.2f}% ({})'.format(sizes_perc[i], sizes[i])
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax2.annotate(percentage, fontstyle = 'italic', fontsize = 16, xy=(x, y), xytext=(1.15*np.sign(x), 1.2*y), horizontalalignment=horizontalalignment, **kw)

ax2.legend(wedges, labels, title="Shift", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize = 14, title_fontsize = 12, facecolor='white')    
    
plt.show()
```
